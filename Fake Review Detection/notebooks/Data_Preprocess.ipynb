## Dataset Description
##### [Source]: https://osf.io/tyue9/
In the endeavor to enhance the reliability of online product reviews, our project involved the creation of a carefully structured dataset consisting of 40,000 entries. Each entry is explicitly labeled as either a Computer Generated Ecommerce Review (CG) or a User Generated Original Review (OR), with a balanced distribution of 20,000 entries in each category.

The selected products cover a wide spectrum of ecommerce offerings, spanning diverse categories. The categories include:
- Home and Kitchen
- Sports and Outdoors
- Electronics
- Movies and TV
- Tools and Home Improvement
- Pet Supplies
- Kindle Store
- Books
- Toys and Games
- Clothing, Shoes, and Jewelry

This curated dataset serves as a valuable resource for researchers and practitioners in the realms of natural language processing and machine learning. Its purpose is to facilitate the development and evaluation of machine learning models specifically designed for the identification and differentiation of authentic user-generated reviews (OR) from computer-generated fake reviews (CG).

The meticulous balance in the distribution of CG and OR entries across diverse product categories aims to capture the intricacies of fake review patterns in various contexts. Leveraging this dataset, the objective is to advance algorithms dedicated to fake review detection, ultimately contributing to the integrity and trustworthiness of ecommerce platforms.
